```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 40)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X, Y, Z)
plt.title('Contour Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import squarify
import matplotlib.cm as cm

sizes = [500, 300, 100, 200]
labels = ['A', 'B', 'C', 'D']
colors = [cm.spring(i / float(len(sizes))) for i in range(len(sizes))]

squarify.plot(sizes=sizes, label=labels, color=colors, alpha=0.8)
plt.title('Treemap')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

categories = ['A', 'B', 'C', 'D']
values = [4, 3, 2, 5]

N = len(categories)
angles = [n / float(N) * 2 * pi for n in range(N)]
values += values[:1]
angles += angles[:1]

plt.polar(angles, values)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Radar Chart')
plt.show()
```
